````python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
````

#Load Dataset

````python
mnist = tf.keras.datasets.mnist
````

````python
(x_train, y_train),(x_test, y_test) = mnist.load_data()
````

````python
print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
````

````python
plt.figure(figsize=(8,8))
for i in range(9):
  plt.subplot(3,3,i+1)
  plt.imshow(x_train[i])
  plt.title(y_train[i])
  plt.axis('off')
````

````python
x_train = tf.keras.utils.normalize(x_train, axis =0)
x_test = tf.keras.utils.normalize(x_test, axis =0)
````

#Building Model

````python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten(input_shape = (28,28)))
model.add(tf.keras.layers.Dense(128, activation = 'relu'))
model.add(tf.keras.layers.Dense(64, activation = 'relu'))
model.add(tf.keras.layers.Dense(32, activation = 'relu'))
model.add(tf.keras.layers.Dense(10, activation='softmax'))
````

````python
model.summary()
````

````python
model.compile(optimizer = 'adam', loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics =['accuracy'])
````

````python
history = model.fit(x_train, y_train, validation_data=(x_test, y_test), batch_size=32, epochs=15)
````

````python
import matplotlib.pyplot as plt

# Plot training & validation loss values
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(loc='upper right')

# Plot training & validation accuracy values
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(loc='lower right')

plt.tight_layout()
plt.show()
````

````python
history1 = model.fit(x_train, y_train, validation_data=(x_test, y_test), batch_size=32, epochs=2)
````

````python
def plot_loss_history(history, title):
    plt.figure(figsize=(8, 4))
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'Model Loss ({title})')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(loc='upper right')
    plt.show()

# Plot the training and validation accuracy
def plot_accuracy_history(history, title):
    plt.figure(figsize=(8, 6))
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'Model Accuracy ({title})')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(loc='lower right')
    plt.show()

# Call the plot functions with the history and corresponding title
plot_loss_history(history1, '2 Epochs')
plot_accuracy_history(history1, '2 Epochs')
````

````python
import matplotlib.pyplot as plt

def plot_loss_histories(histories, titles):
    plt.figure(figsize=(8, 6))

    for history, title in zip(histories, titles):
        plt.plot(history.history['loss'], label=f'Training Loss ({title})')
        plt.plot(history.history['val_loss'], label=f'Validation Loss ({title})')
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(loc='upper right')
    plt.show()

def plot_accuracy_histories(histories, titles):
    plt.figure(figsize=(8, 6))

    for history, title in zip(histories, titles):
        plt.plot(history.history['accuracy'], label=f'Training Accuracy ({title})')
        plt.plot(history.history['val_accuracy'], label=f'Validation Accuracy ({title})')
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(loc='lower right')
    plt.show()

# Call the plot functions with the histories and corresponding titles
plot_loss_histories([history1, history2], ['2 Epochs', '8 Epochs'])
plot_accuracy_histories([history1, history2], ['2 Epochs', '8 Epochs'])
````
